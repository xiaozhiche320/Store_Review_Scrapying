```python

from google_play_scraper import Sort, reviews

app_id = 'com.babbel.mobile.android.en'  # 替换为你想抓取评论的应用的 id

result, continuation_token = reviews(
    app_id,
    lang='en',  # 设定评论语言，这里设定为英语
    country='us',  # 设定国家，这里设定为美国
    sort=Sort.MOST_RELEVANT,  # 评论的排序方式，这里选择最相关
    count=500 # 抓取评论的数量，这里选择抓取200条评论
)

# result 是一个包含评论的列表，你可以将它转换为 DataFrame

import pandas as pd

df = pd.DataFrame(result)
df.head()
```

```python
# Parse the 'date' column to datetime format
df['date'] = pd.to_datetime(df['at'])
df = df.sort_values(by='at', ascending=False)
df.head()
```

```python
df = df.replace({r'[^\x00-\x7F]+':''}, regex=True)
```

```python
from textblob import TextBlob

# Function to apply sentiment analysis


def get_sentiment(text):
    return TextBlob(text).sentiment.polarity

# Add sentiment analysis column
df['sentiment'] = df['content'].apply(get_sentiment)


# Filter reviews with negative sentiment
df_negative = df[df['sentiment'] < 0]
# Write the negative reviews to a CSV file
df_negative.to_csv(r'C:\Users\WANGYILIN\Desktop\sister_program\BABBEL_google-negative-reviews.csv', encoding='utf-8', index=False)
```

```python
df_positive = df[df['sentiment'] > 0]
df_positive.to_csv(r'C:\Users\WANGYILIN\Desktop\sister_program\BABBEL_google-positive-reviews.csv', encoding='utf-8', index=False)
```
