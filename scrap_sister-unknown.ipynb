```python
import pandas as pd
import numpy as np
import json


from app_store_scraper import AppStore
slack = AppStore(country='us', app_name='duolingo-language-lessons', app_id = '570060128')

slack.review(how_many=500)
```

```python
slack.reviews
```

```python
slackdf = pd.DataFrame(np.array(slack.reviews),columns=['review'])
slackdf2 = slackdf.join(pd.DataFrame(slackdf.pop('review').tolist()))
slackdf2.head()
```

```python
# Parse the 'date' column to datetime format
slackdf2['date'] = pd.to_datetime(slackdf2['date'])
slackdf2 = slackdf2.sort_values(by='date', ascending=False)
slackdf2.head()
```

```python
slackdf2 = slackdf2.replace({r'[^\x00-\x7F]+':''}, regex=True)
```

```python
slackdf2.to_csv(r'C:\Users\WANGYILIN\Desktop\sister_program\Duolingo.csv' ,encoding='utf-8', index=False)
```

```python
import nltk
from textblob import TextBlob
```

```python
# Function to apply sentiment analysis


def get_sentiment(text):
    return TextBlob(text).sentiment.polarity

# Add sentiment analysis column
slackdf2['sentiment'] = slackdf2['review'].apply(get_sentiment)


# Filter reviews with negative sentiment
df_negative = slackdf2[slackdf2['sentiment'] < 0]
# Write the negative reviews to a CSV file
df_negative.to_csv(r'C:\Users\WANGYILIN\Desktop\sister_program\Duolingo-negative-reviews.csv', encoding='utf-8', index=False)

```

```python
df_positive = slackdf2[slackdf2['sentiment'] > 0]
df_positive.to_csv(r'C:\Users\WANGYILIN\Desktop\sister_program\Duolingo-positive-reviews.csv', encoding='utf-8', index=False)
```
